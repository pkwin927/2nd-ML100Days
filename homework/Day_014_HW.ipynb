```python
# 載入需要的套件

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')
```

```python
app_train = pd.read_csv("C:/Users/User/Desktop/100Days/application_train.csv")
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
housetype_unique = app_train["HOUSETYPE_MODE"].unique()

nrows = len(housetype_unique) //2 +1
ncols = 2

plt.figure(figsize=(20,30))
for i in range(len(housetype_unique)):
    plt.subplot(nrows, ncols, i+1)
    app_train.loc[app_train['HOUSETYPE_MODE']==housetype_unique[i], 'AMT_CREDIT'].hist() 
    plt.ylabel(str(housetype_unique[i]))
    plt.xlabel('AMT_CREDIT')
plt.show()
```
