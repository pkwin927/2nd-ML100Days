```python
from sklearn import datasets
from sklearn.model_selection import cross_val_predict
from sklearn import tree
from sklearn import linear_model
from sklearn import ensemble
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

DT = tree.DecisionTreeRegressor()
RF = ensemble.RandomForestRegressor()

boston = datasets.load_boston()

y = boston.target

y = y.reshape(-1,1)
```

```python
predicted = cross_val_predict(DT, boston.data, y, cv=10)



DT.fit(boston.data, y)

predict_y=DT.predict(boston.data[2].reshape(1,-1))

plt.scatter(predicted,y,s=2)

plt.plot(predict_y, predict_y, 'ro')

plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)

plt.xlabel('Predicted')

plt.ylabel('Measured')
```

```python
predicted = cross_val_predict(RF, boston.data, y, cv=10)



RF.fit(boston.data, y)

predict_y=RF.predict(boston.data[2].reshape(1,-1))

plt.scatter(predicted,y,s=2)

plt.plot(predict_y, predict_y, 'ro')

plt.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=2)

plt.xlabel('Predicted')

plt.ylabel('Measured')
```
