```python
"""
Your Code Here
"""
matrix = np.random.uniform(-1, 1, (10,10))

plt.figure(figsize=(10,10))

heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
plt.title('Correlation Heatmap-Random');
```

```python
nrow = 1000
ncol = 3

matrix = np.random.uniform(-1,1, (nrow,ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)
plot_data["target"] = plot_data.index

grid = sns.PairGrid( data = plot_data , hue= "target" , size = 3, vars= [0 , 1 ,2], diag_sharey= False)

grid.map_upper(plt.scatter , alpha = 0.2 )
grid.map_diag(sns.distplot )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)
```

```python
nrow = 1000
ncol = 3

matrix = np.random.normal(size = (nrow,ncol))
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

plot_data["target"] = plot_data.index

grid = sns.PairGrid( data = plot_data , hue= "target" , size = 3, vars= [0 , 1 ,2], diag_sharey= False)
grid.map_upper(plt.scatter , alpha = 0.2 )
grid.map_diag(sns.distplot )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
